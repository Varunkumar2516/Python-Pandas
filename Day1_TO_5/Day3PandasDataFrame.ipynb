# Selecting Rows From Dataframe

- __.iloc[] :Integer-location based ,(Integer positions (0, 1, 2, ...))__

__(similar to List access)__

__df.iloc[row_index, col_index]__

- __.loc[] :Label-based access ,(Labels (index/column names))__

__(similar to Dictionary Access)__

__df.loc[row_label, col_label]__

```python
import pandas as pd

data = {
    'index':['a', 'b', 'c', 'd'],
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'City': ['NY', 'Paris', 'London', 'Tokyo']
}

Dataframe = pd.DataFrame(data)
Dataframe.set_index('index',inplace=True)
print(Dataframe)
```

# __iloc[rowindex,Colindex]__

```python
"""iloc does not need of any label (index ) it just need of integer value 
may be negative same as how we access the things in list
 -2-> 2ndLast Element
 -1-> last Element
 0 -> first element
 1-> 2nd element and so on
  
  this very simple same as list(use the Positions that are inbuilt(internally) defined
   by python when object created (..........-2,-1,0,1,2,3..........))

  
During Slicing THe end posititon is Excluded From SLicing 
 """
```

```python
Dataframe.iloc[0]
```

```python
Dataframe.iloc[1]
```

```python
Dataframe.iloc[-1]
```

```python
Dataframe.iloc[::-1] # reverse
```

```python
# Slicing :ENd position exluded
Dataframe.iloc[1:2] # 2 is Excluded
```

```python
Dataframe.iloc[1::2]
```

```python
"""Using Fancy Indexing with .iloc[]"""
# Fetch only first 2 rows and two columns using fancy slicing
print(Dataframe.iloc[[0,1],[0,1]])
# fetch 1st and last row and 2 and 1st column in same order
print(Dataframe.iloc[[0,-1],[1,0]])
```

# __loc[row_label ,col_label]__

```python
"""
.loc[] is label-based indexing 
— it uses the actual names or labels of rows and columns, not positions.

- Negative INdexing Not works (until we manually define the index in our Dataframe negative )
-1 -> not works
-2 -> not works 
if we manually create the index that have negative index than we can use its label

here only the labels are used not the positions 
-> default labels are starts from (0 -> so on) so if Dataframe is created with  default index
-> we can use the labels like 
[0]-> first
[1]-> second
[2] -> third 
so on

-> and if we manually define the Index 
-> we have to use those indexes
['a'] first 
['b'] second 
['c'] third
so on


-> During Slicing the endlabel IN included in Slicing 

Summaary
.loc does NOT use integer positions — it uses **labels** of rows and columns.
- You must use actual names/index values like 'a', 'b', 'Name', 'Age'.
- Negative numbers wont work unless your index/column labels are actually negative.
- Its more readable and is used when working with named rows/columns.

So:
  df.loc['a']        → accesses row labeled 'a'
  df.loc['b', 'Age'] → accesses b Age
  df.loc['a':'b']    → includes both 'a' and 'b' (inclusive slicing)

"""
```

```python
Dataframe
```

```python
Dataframe.loc['a']      # using the index label
```

```python
Dataframe.loc['b']
```

```python
Dataframe.loc['d']
```

```python
""" as We see the lables are defined manually as a,b,c,d 
so if we try to access elements with index Position it will throw error"""

# Dataframe.loc[0]  Error!!!!!!!!
```

```python
# slicing 
Dataframe.loc['a':'c']   # c in included
```

```python
#USing column name also
Dataframe.loc['a',"Name"] 
```

```python
# get only name
Dataframe.loc[:,'Name']
```

```python
Dataframe.loc['a':'c','Name':"Age"]   # c,Age in included
```

```python
# Reverse the Dataframe
Dataframe.loc[::-1,::-1]
```

```python
"""Using Fancy Indexing with .iloc[]"""
# Fetch only first 2 rows and two columns using fancy slicing
print(Dataframe.loc[['a','b'],['Name','Age']])
# fetch 1st and last row and 2 and 1st column in same order
print(Dataframe.loc[['a','d'],['Age','Name']])
```

# Filtering Data in Pandas (Boolean Indexing)

__Filtering means selecting rows that meet certain conditions__

```python
df
```

1. Basic Filtering (Single Condition)

```python
df[df['age'] > 30]
```

2. Multiple Conditions

```python
df[(df['age'] > 15) & (df['weight'] >40)]
```

3. Filtering with .isin()

```python
df[df['Name'].isin(['Harsh', 'prem'])]
```
