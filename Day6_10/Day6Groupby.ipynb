__Groupby__:

__we use groupby() to split data into groups (Categorical COlumns), perform operations on each group independently, and then combine the results. This is especially useful for data aggregation, transformation, and analysis.__

__The main reason is to analyze data by categories or groups. Think of it as:__
__Split → Apply → Combine__

__When a column (say "Team" or "Category column") has repeated values, groupby() lets you collapse those repeats into a single group, and then you can perform operations like sum(), mean(), max() on the other columns within each group.__

__We mostly Apply  Group by on Categorical COlumns__

```python
"""
Think of it like:
Team	  Score
A	       10
A	       20
B	       15
B	       25
C	       30
"""
"groupby lets you first combine these Team columns and Perform Operations on Other COlumns(Score)"
"like sum,mean,max"
"df.groupby('Team')['Score'].sum()"
" this syntax is used "



"""We get like below
Team	Score (sum)
A	      30
B	      40
C	      30
"""
```

# __Group By Attributes and Method__

```python
"""sum()	Total per group

mean()	Average per group

count()	Non-null count per group

size()	Total count (including nulls)

max() / min()	Max / Min value per group

agg()	Multiple aggregations

get_group('name')	Access data for one group

groups	Shows group keys and row indices

apply()	Apply any custom function to each group

transform()	Add results back to original DataFrame shape"""
```

- 12. __agg()__

__.agg() stands for aggregate. It lets you:__

__Apply one or multiple functions to one or more columns after grouping.Customize the output, including column names.__

```python
# 1 single function apply using agg() : pass the functions simply :
#  use ("method")
movies_df.groupby("Genre")['Gross'].agg("sum")
```

```python
#  multiple Functions apply using agg() :use list of functions:
# ['method1','methd2']


movies_df.groupby("Genre")['Gross'].agg(["sum",'mean','max','min'])
```

```python
# applying different Functions on DIfferent COlumns using agg(): 
# pass DIctionary like
# {"Columns" : "method"}

"CUstom  Column Names (With Dict)"

movies_df.groupby("Genre").agg({
   "Runtime":"mean",
   "IMDB_Rating" :"mean",
   'No_of_Votes':"sum",
   'Metascore':"min",
   'Gross':"sum"
   })
```

```python
#  multiple different Functions apply on DIfferent COlumns using agg() : use dictionary of
#  {"Column":['method1','method2']}
movies_df.groupby("Genre").agg({
   "Runtime":["mean",'min'],
   "IMDB_Rating" :["mean",'max'],
   'No_of_Votes':["sum",'mean'],
   'Metascore':["min",'max'],
   'Gross':"sum"
   })
```

```python
for name,dataframe in Mygroup:
    RequiredData=['Comedy','Acion']
    if name in RequiredData:
        print(f"Department: {name}")
        print(dataframe.head(1))
        print("---")

# wecan perform any custom logic ussing the For looops on Groupby objects
```

__Why loop like this?__
__WE can:__

**Apply different logic for each group**

**Do custom calculations, transformations**

**Write group-specific files, plots, summaries, etc.**
